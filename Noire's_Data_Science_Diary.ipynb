**My First Impression Pre-Run **
Spam_amount = 0 so first thing outputted is 0

Then we add 0 + 4

Then in our if statement, if spam_amount > 0 (which it is) we print "But I don't want ANY spam!"

Outside of that if statement we print viking song which is "Spam" * spam_amount, which prints the string "Spam" 4 times

```python
spam_amount = 0
print(spam_amount)

# Ordering Spam, egg, Spam, Spam, bacon and Spam (4 more servings of Spam)
spam_amount = spam_amount + 4

if spam_amount > 0:
    print("But I don't want ANY spam!")

viking_song = "Spam " * spam_amount
print(viking_song)
```

Pretty cool right

You might be wondering how "Spam" * 4 = "Spam" "Spam" "Spam" "Spam"

This is a case of Python's [**Operator Overloading**](https://https://en.wikipedia.org/wiki/Operator_overloading) where operators hold different functions depending on what their arguements are.

i.e. "Hello" + "World" would output "HelloWorld", but 4 + 4 = 8, not 44

```python
Hi = "Hello"
Earth = "World"
print(Hi + Earth)

4 + 4
```
