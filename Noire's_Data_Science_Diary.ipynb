Moving on, I'm pressuming we all know how to define a function in python, so I'll cover some other things pertaining to functions

-> Docstrings, If you've done Leetcode you've seen these, but to recap this is how we can write the description of a function in plain english


```python
#Code from kaggle

def least_difference(a, b, c):
    """ This is our Docstring, i.e. a triple-quoted string

    Return the smallest difference between any two numbers
    among a, b and c.

    >>> least_difference(1, 5, -5)
    4
    """
    diff1 = abs(a - b)
    diff2 = abs(b - c)
    diff3 = abs(a - c)
    return min(diff1, diff2, diff3)

```

Wanna see a cool trick? Lets revist that help function again and see what it thinks of our function.

```python
help(least_difference)
```

Conceptually, you should be able to read functions which operate within other functions. When we approach Machine Learning & Stats this becomes extremely important

```python
#code from kaggle
def mult_by_five(x):
    return 5 * x

def call(fn, arg):
    """Call fn on arg"""
    return fn(arg)

def squared_call(fn, arg):
    """Call fn on the result of calling fn on arg"""
    return fn(fn(arg))

print(
    call(mult_by_five, 1),
    squared_call(mult_by_five, 1),
    sep='\n', # '\n' is the newline character - it starts a new line
)
```

**Series**: a one-dimensional array holding data of any type.

```python
a = [1, 7, 2]

myvar = pd.Series(a)

print(myvar)
```

Notice how theres a 0, 1, 2 column auto generated?
That's our index number, which is auto generated when we haven't assigned **labels**

We can also call our values by their respective indices

```python
print(myvar[0])
```

Now lets create some labels with the *index* arguement

```python
a = [1, 7, 2]

myvar = pd.Series(a, index = ['First', 'Second', 'Third'])

print(myvar)
```

Similarly, we can still return the value at the index with labels


```python
print(myvar['First'])
```

Like a dictionary, **we can use key:value objects** when working with Series in Pandas

```python
calories = {"day1": 420, "day2": 380, "day3": 390}

myvar = pd.Series(calories)

print(myvar)
```

Additionally we can use our index function to only return specific items in our series

```python
alories = {"day1": 420, "day2": 380, "day3": 390}

myvar = pd.Series(calories, index = ["day3"])

print(myvar)
```
